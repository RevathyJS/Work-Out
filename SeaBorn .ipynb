```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

# Correlation with Heatmap

A correlation heatmap uses colored cells, typically in amonochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in feature selection.

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join Plot

A join plot allows to study the relationship between 2 umeris variables. The central chart display their correlation. it is usually a scatterplot, a hexbin plot, a 2D density plot

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot 

A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's vlaue like this. Pair plot are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

Dist plot helps to check the distribution ofthe columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorial Plots

Seaborn also helps in doing the analysis on categorial data points. 

- Boxplot
- violinplot
- count plot
- bar plot

```python
sns.countplot('sex',data=df)
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
df.head()
```

# Box plot
 A box and whisker plot(sometimes called a boxplot) is a graph that presents information from five number summary.

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violin Plot

Violin plot helps to see both the distribution of data in terms of kernel density estimation and the box plot.


```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
